```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F

x = np.linspace(-5, 5, 1000)

# Sigmoid
y_sigmoid = 1 / (1 + np.exp(-x))

# Hyperbolic Tangent
y_tanh = ( 2 / (1 + np.exp(-2*x) ) ) -1

# Relu
y_relu = np.maximum(0, x)

# Leaky Relu
def prelu(x):
    activations = []
    for value in x:
        if value>0 :
            activations.append(value)
        else:
            activations.append(0.1*value)
    return activations
    
y_prelu = prelu(x)

# Swish and Mish
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x, beta=1):
    return (x * sigmoid(beta * x))

def mish(x):
    return torch.tensor(x)*torch.tanh(F.softplus(torch.tensor(x)))

y_swish = swish(x)
y_mish = mish(x)

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
x_ticks = [-5, -2.5, 0, 2.5, 5]
y_ticks = [-1, 0, 1, 2]
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
print(ax.get_ylim())
ax.set_ylim(ax.get_ylim()[0],2)
plt.plot(x,y_sigmoid, linewidth=3.0)
plt.plot(x,y_tanh, linewidth=3.0, color="green")
plt.plot(x,y_relu, linewidth=3.0,color="red")
plt.plot(x,y_prelu, linewidth=3.0, color="mediumturquoise")
plt.plot(x,y_swish, linewidth=2.0, color="coral")
plt.plot(x,y_mish, linewidth=2.0, color="lawngreen")
fig.legend(['Sigmoid','Tanh','ReLU','PReLU','Swish','Mish'], prop={"size":12}, loc="upper right", bbox_to_anchor=(0.35,0.82))
plt.show()
fig.savefig("Activations_2.pdf")


```
